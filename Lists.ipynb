```
#List Average
def main():
  # Set up variables
  sum = 0.0
  NUMBER_OF_ENTRIES = 5
  numbers = []
  # Get input from user
  print("Please enter", NUMBER_OF_ENTRIES, "numbers: ")
  for i in range(0, NUMBER_OF_ENTRIES):
    num = eval(input("Enter number " + str(i) + ": "))
    numbers += [num]
    sum += num;
    # Print the numbers entered
    print(end="Numbers entered: ")
  for num in numbers:
    print(num, end=" ")
    print()
  print("Average:", sum/NUMBER_OF_ENTRIES)

main()
```

```
def sum(lst):
  result = 0;
  for item in lst:
    result += item
  return result

def clear(lst):
  for i in range(len(lst)):
    lst[i] = 0

def random_list(n):
#Builds a list of n integers, where each integer is a pseudorandom number in the range 0...99. Returns the random list.
  import random
  result = []
  for i in range(n):
    rand = random.randrange(100)
    result += [rand]
  return result

def main():
  a = [2, 4, 6, 8]
  # Print the contents of the list
  print(a)
  # Compute and display sum
  print(sum(a))
  # Zero out all the elements of list
  clear(a)
  # Reprint the contents of the list
  print(a)
  # Compute and display sum
  print(sum(a))
  # Test empty list
  a = []
  print(a)
  print(sum(a))
  # Test pseudorandom list with 10 elements
  a = random_list(10)
  print(a)
  print(sum(a))

main()
```

# List Comprehension

```
# find the squares of a number using the for loop:
numbers = [1, 2, 3, 4]
squares = []

for n in numbers:
  squares.append(n**2)

print(squares)  # Output: [1, 4, 9, 16]
```

```
# find the squares of a number using the list Comprehension
numbers = [1, 2, 3, 4]
squares = [n**2 for n in numbers]

print(squares)  # Output: [1, 4, 9, 16]
```

```
#Find common numbers from two lists using for loop.
list_a = [1, 2, 3, 4]
list_b = [2, 3, 4, 5]

common_num = []

for a in list_a:
  for b in list_b:
    if a == b:
      common_num.append(a)
      
print(common_num)  # Output [2, 3, 4]
```

```
#Find common numbers from two lists using list Comprehension
list_a = [1, 2, 3, 4]
list_b = [2, 3, 4, 5]

common_num = [a for a in list_a for b in list_b if a == b]

print(common_num) # Output: [2, 3, 4]
```

```
'''
Q1. Write a program to perform following tasks:
- Read a list of integers from a user.
- Remove duplicate values from this list. We call it a super list.
- In a single scan of the super list, perform follwing operations:
    - If the number is not perfectly divisible by 3, discard it.
    - If the number is greater than 30, add 5 to it. Else, subtract 5 from it.
- Display contents of the list.
'''
input_string = input("Enter a list elements separated by space ")
userList = input_string.split()
superlist= list(set(userList))
print(superlist)
i=0
for el1 in superlist:
  el=int(el1)
  if el%3!=0:
    superlist.remove(el1)
    continue
  if el>30:
    superlist[i]=el+5
  else:
    superlist[i]=el-5
  i=i+1
print(superlist)
```
