# String Basics

```
#Creating String
my_string = 'Hello'
print(my_string)

my_string = "Hello"
print(my_string)

my_string = '''Hello,
    my name is bhavana'''
print(my_string)

# triple quotes string can extend multiple lines
my_string = """Hello, welcome to
           the world of Python"""
print(my_string)
```

```
#access characters in a string
"""We can access individual characters using indexing and a range of characters using slicing. 
Index starts from 0. Trying to access a character out of index range will raise an IndexError. 
The index must be an integer. Python allows negative indexing for its sequences. 
The index of -1 refers to the last item. We can access a range of items in a string by using the slicing operator : """

#Accessing string characters in Python
str = 'Python Programming'
print('str = ', str)

#first character
print('str[0] = ', str[0])

#last character
print('str[-1] = ', str[-1])

#slicing 2nd to 5th character
print('str[1:5] = ', str[1:5])

#slicing 6th to 2nd last character
print('str[5:-2] = ', str[5:-2])
```

```
#change or delete a string
"""Strings are immutable. This means that elements of a string cannot be changed once they have been assigned."""
str[5] = 'a'
#you can change entire string for object or delete whole string using del
```

```
#String Operations
str1 = 'Hello'
str2 ='World!'

# using +
print('str1 + str2 = ', str1 + str2)

# using *
print('str1 * 3 =', 3*str1)
```

```
#Iterating Through a string
count = 0
for letter in 'Hello World':
    if(letter == 'l'):
        count += 1
print(count,'letters found')
```

```
#String Membership Test
print('a' in 'program')
print('ate' not in 'battle')
```

```
word1 = 'Wow'
word2 = 'Wow'
print('Equality:', word1 == word2, ' Alias:', word1 is word2)
word1 = [1,2,3]
word2 = [1,2,3]
print('Equality:', word1 == word2, ' Alias:', word1 is word2)
```

# String formatting

```
print("This {} is a good option for beginners in python".format("tutorial"))
my_string = "This {} is a good option for beginners in python"
print(my_string.format("tutorial"))
```

```
my_string = "This {} is a good option for beginners in {}"
print(my_string.format("tutorial","Machine Learning"))
```

```
my_string = "This {1} is a good option for beginners in {0}"
domain_val="Machine Learning"
print(my_string.format("tutorial","Machine Learning"))
my_string = "This {0} is a good option for beginners in {domain}"
print(my_string.format("tutorial",domain=domain_val))
```

```
my_string = "The Temperature in {0} today is {1:d} degrees outside!"
print(my_string.format("Vizag",22))
my_string = "The Temperature in {0} today is {1:.2f} degrees outside!"
print(my_string.format("Vizag",22.025))
```

```
my_string = "The Temperature in {0:20} today is {1:d} degrees outside!"
print(my_string.format("Vizag",22))
for i in range(4,15):
    print("{:6d} {:6d} {:6d}".format(i,i*i,i*i*i))

```
